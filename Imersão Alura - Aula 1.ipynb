```python
import yfinance as yf

ticker = "JPM"
start_date = "2021-01-01"
end_date = "2025-07-31"

jpm_data = yf.download(ticker, start=start_date, end=end_date)

# Exibir os primeiros registros
display(jpm_data.head())

# Exibir os últimos registros
display(jpm_data.tail())
```

```python
# Conte a quantidade de registros NaN na coluna 'Close_JPM'
nan_count_close_jpm = jpm_data['Close_JPM'].isna().sum()

# Imprima o resultado
print(f"Número de registros NaN na coluna 'Close_JPM': {nan_count_close_jpm}")
```

```python
# Filtrar o DataFrame para incluir apenas registros do ano de 2025
jpm_data_2025 = jpm_data[jpm_data.index.year => 2025]

# Calcular o 'Daily Return' para os dados de 2025
jpm_data_2025['Daily Return'] = jpm_data_2025['Close_JPM'].pct_change()

# Exibir os primeiros registros do DataFrame filtrado com o 'Daily Return' calculado
display(jpm_data_2025.head())
```

```python
import matplotlib.pyplot as plt

# Remover linhas com valores NaN na coluna 'Daily Return' antes de plotar (o primeiro valor é NaN)
jpm_data_2025_cleaned = jpm_data_2025.dropna(subset=['Daily Return'])

plt.figure(figsize=(12, 6))
plt.scatter(jpm_data_2025_cleaned.index, jpm_data_2025_cleaned['Daily Return'], alpha=0.5, s=10)
plt.title('Daily Returns Dispersion of JPM Stock in 2025')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filtrar o DataFrame para incluir apenas registros dos anos de 2023, 2024 e 2025
jpm_data_2023_2025 = jpm_data[jpm_data.index.year.isin([2023, 2024, 2025])].copy()

# Calcular o 'Daily Return' para os dados filtrados
jpm_data_2023_2025['Daily Return'] = jpm_data_2023_2025['Close_JPM'].pct_change()

# Remover linhas com valores NaN na coluna 'Daily Return' antes de plotar
jpm_data_2023_2025_cleaned = jpm_data_2023_2025.dropna(subset=['Daily Return'])

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotar Daily Return no eixo Y esquerdo (primário)
ax1.scatter(jpm_data_2023_2025_cleaned.index, jpm_data_2023_2025_cleaned['Daily Return'], alpha=0.5, s=5, label='Daily Return')
ax1.set_xlabel('Date')
ax1.set_ylabel('Daily Return', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True)

# Criar um segundo eixo Y para o preço de fechamento
ax2 = ax1.twinx()

# Plotar Close Price no eixo Y direito (secundário)
ax2.plot(jpm_data_2023_2025.index, jpm_data_2023_2025['Close_JPM'], color='tab:red', label='Close Price')
ax2.set_ylabel('Close Price', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adicionar título
plt.title('Daily Returns Dispersion and Close Price of JPM Stock (2023-2025)')

# Adicionar legendas (opcional, pois os eixos já estão coloridos)
# fig.legend(loc="upper left")

plt.show()
```
